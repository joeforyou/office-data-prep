```python
import pandas as pd

df = pd.read_csv('the-office-lines.csv')
```

```python
import nltk
import matplotlib
import matplotlib.pyplot as plt

#nltk.download('punkt')

top_N = 10

#m = df[df.speaker == "Michael"]
 
txt = df.line_text.str.lower().str.replace(r"(\[.*\])",' ').str.cat(sep=' ')
words = nltk.tokenize.word_tokenize(txt)
```

```python
words
```

```python
#nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import wordpunct_tokenize

word_dist = nltk.FreqDist(words)
stopwords = set(stopwords.words('english'))
stopwords.update(['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}']) # remove it if you need punctuation 
stopwords.update(["'re","'s","n't","'m","..."])

words_except_stop_dist = nltk.FreqDist(w for w in words if w not in stopwords)
```

```python
stopwords
```

```python
print('All Frequencies, including stopwords:')
print('=' * 60)
rlst = pd.DataFrame(word_dist.most_common(top_N),
                   columns=['word','frequency'])
print(rlst)
```

```python
rlst = pd.DataFrame(words_except_stop_dist.most_common(top_N),
                   columns=['word','frequency']).set_index('word')

matplotlib.style.use('ggplot')

rlst.plot.bar(rot=0)
```

```python
rlst = pd.DataFrame(words_except_stop_dist.most_common(top_N),
                   columns=['word','frequency'])

print(rlst)
```
